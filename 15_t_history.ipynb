## Global Temperature history

Let's upload two files: deep_sea_sediment.txt and domeC.txt.

```python
import numpy as np
import matplotlib.pyplot as plt

# Read d18O from deep ocean sediment
data=np.loadtxt('deep_sea_sediment.txt')
time=data[:,0]
d18o=data[:,1]
```

#### Exercise

Plot d18o from deep see sediment (x-axis: time, y-axis: d18o)

```python
# Read dD from DomeC ice core
data2=np.loadtxt('domeC.txt')
time2=data2[:,4]/1000.# time in thousand years
delD=data2[:,5]
```

```python
# two y axis
# axis 1
fig,ax1=plt.subplots()
ax1.plot(time,d18o,'b')
ax1.set_ylabel('Ocean d18O',color='b')
ax1.tick_params('y', colors='b')
plt.xlabel('kyr before present')
plt.ylim(6,2)

# axis 2
ax2=ax1.twinx()
ax2.plot(time2,delD,'r')
ax2.set_ylabel('Dome C delD',color='r')
ax2.tick_params('y', colors='r')
plt.xlim(0,500)
plt.show()
```
